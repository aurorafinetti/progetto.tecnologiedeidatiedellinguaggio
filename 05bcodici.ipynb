```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML

# ── DATASET STORICO VENDITE VINILE USA (fonte: RIAA) ──────────────────
dati_riaa = [
    {"anno": 2005, "unita_milioni": 0.9,  "ricavi_milioni": 14.2},
    {"anno": 2006, "unita_milioni": 0.9,  "ricavi_milioni": 15.6},
    {"anno": 2007, "unita_milioni": 1.3,  "ricavi_milioni": 20.8},
    {"anno": 2008, "unita_milioni": 1.9,  "ricavi_milioni": 31.8},
    {"anno": 2009, "unita_milioni": 2.5,  "ricavi_milioni": 41.8},
    {"anno": 2010, "unita_milioni": 2.8,  "ricavi_milioni": 55.8},
    {"anno": 2011, "unita_milioni": 3.9,  "ricavi_milioni": 87.0},
    {"anno": 2012, "unita_milioni": 4.6,  "ricavi_milioni": 107.0},
    {"anno": 2013, "unita_milioni": 6.1,  "ricavi_milioni": 146.0},
    {"anno": 2014, "unita_milioni": 9.2,  "ricavi_milioni": 226.0},
    {"anno": 2015, "unita_milioni": 11.9, "ricavi_milioni": 416.0},
    {"anno": 2016, "unita_milioni": 13.1, "ricavi_milioni": 422.0},
    {"anno": 2017, "unita_milioni": 14.3, "ricavi_milioni": 459.0},
    {"anno": 2018, "unita_milioni": 16.8, "ricavi_milioni": 497.0},
    {"anno": 2019, "unita_milioni": 18.8, "ricavi_milioni": 504.0},
    {"anno": 2020, "unita_milioni": 27.5, "ricavi_milioni": 626.0},
    {"anno": 2021, "unita_milioni": 41.7, "ricavi_milioni": 1017.0},
    {"anno": 2022, "unita_milioni": 43.0, "ricavi_milioni": 1224.0},
    {"anno": 2023, "unita_milioni": 43.4, "ricavi_milioni": 1285.0},
    {"anno": 2024, "unita_milioni": 46.8, "ricavi_milioni": 1400.0},
    {"anno": 2025, "unita_milioni": 46.8, "ricavi_milioni": 1000.0},
]

df = pd.DataFrame(dati_riaa)
df.to_csv("riaa_vinyl_storico.csv", index=False)
print("✅ Dataset creato con successo!")

# ── TABELLA COMPLETA ──────────────────────────────────────────────────
righe_html = ""
colore_alt = False
for _, r in df.iterrows():
    bg = "#f9f9f9" if colore_alt else "#ffffff"
    colore_alt = not colore_alt
    righe_html += (
        f"<tr style='background:{bg}'>"
        f"<td style='padding:5px 12px; text-align:center'><b>{r['anno']}</b></td>"
        f"<td style='padding:5px 12px; text-align:center'>{r['unita_milioni']}</td>"
        f"<td style='padding:5px 12px; text-align:center'>${r['ricavi_milioni']}M</td>"
        f"</tr>"
    )

display(HTML(f"""
<h3>📊 Vendite di vinili negli USA – 2005/2025 (fonte: RIAA)</h3>
<p style='color:gray'>Totale anni analizzati: <b>{len(df)}</b></p>
<div style='max-height:500px; overflow-y:auto'>
<table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px; width:50%'>
  <thead style='background:#1a1a2e; color:white; position:sticky; top:0'>
    <tr>
      <th style='padding:7px 12px'>Anno</th>
      <th style='padding:7px 12px'>Unità vendute (milioni)</th>
      <th style='padding:7px 12px'>Ricavi (milioni $)</th>
    </tr>
  </thead>
  <tbody>{righe_html}</tbody>
</table>
</div>
"""))

# ── VARIAZIONE ANNO SU ANNO ───────────────────────────────────────────
df["variazione_%"] = df["unita_milioni"].pct_change().mul(100).round(1)

righe_var = ""
colore_alt = False
for _, r in df.iterrows():
    bg = "#f9f9f9" if colore_alt else "#ffffff"
    colore_alt = not colore_alt
    var = r["variazione_%"]
    if pd.isna(var):
        colore_var = "#888"
        testo_var = "–"
    elif var > 0:
        colore_var = "#1db954"
        testo_var = f"+{var}%"
    else:
        colore_var = "#e85d4a"
        testo_var = f"{var}%"

    righe_var += (
        f"<tr style='background:{bg}'>"
        f"<td style='padding:5px 12px; text-align:center'><b>{r['anno']}</b></td>"
        f"<td style='padding:5px 12px; text-align:center'>{r['unita_milioni']}M</td>"
        f"<td style='padding:5px 12px; text-align:center; color:{colore_var}; font-weight:bold'>{testo_var}</td>"
        f"</tr>"
    )

display(HTML(f"""
<h3>📈 Crescita anno su anno</h3>
<table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px; width:50%'>
  <thead style='background:#1a1a2e; color:white'>
    <tr>
      <th style='padding:7px 12px'>Anno</th>
      <th style='padding:7px 12px'>Unità vendute</th>
      <th style='padding:7px 12px'>Variazione %</th>
    </tr>
  </thead>
  <tbody>{righe_var}</tbody>
</table>
"""))
```
