```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML
from collections import defaultdict

df = pd.read_csv("TikTok_songs_2022.csv")

# ── TABELLA COMPLETA ──────────────────────────────────────────────────
def genera_tabella(dataframe, titolo=""):
    righe_html = ""
    colore_alt = False
    for _, r in dataframe.iterrows():
        bg = "#f9f9f9" if colore_alt else "#ffffff"
        colore_alt = not colore_alt
        righe_html += (
            f"<tr style='background:{bg}'>"
            f"<td style='padding:5px 12px'>{r['track_name']}</td>"
            f"<td style='padding:5px 12px'>{r['artist_name']}</td>"
            f"<td style='padding:5px 12px'>{r['album']}</td>"
            f"<td style='padding:5px 12px; text-align:center'>{r['track_pop']}</td>"
            f"<td style='padding:5px 12px; text-align:center'>{round(r['danceability'], 2)}</td>"
            f"<td style='padding:5px 12px; text-align:center'>{round(r['energy'], 2)}</td>"
            f"</tr>"
        )
    return HTML(f"""
    <h3>🎵 {titolo}</h3>
    <p style='color:gray'>Totale: <b>{len(dataframe)}</b> brani</p>
    <div style='max-height:500px; overflow-y:auto'>
    <table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px; width:100%'>
      <thead style='background:#ff0050; color:white; position:sticky; top:0'>
        <tr>
          <th style='padding:7px 12px'>Titolo</th>
          <th style='padding:7px 12px'>Artista</th>
          <th style='padding:7px 12px'>Album</th>
          <th style='padding:7px 12px'>Popolarità</th>
          <th style='padding:7px 12px'>Danceability</th>
          <th style='padding:7px 12px'>Energia</th>
        </tr>
      </thead>
      <tbody>{righe_html}</tbody>
    </table>
    </div>
    """)

display(genera_tabella(df, "Brani virali su TikTok – 2022"))

# ── TOP 10 BRANI PIÙ POPOLARI ─────────────────────────────────────────
top10 = df.nlargest(10, "track_pop")[["track_name", "artist_name", "track_pop"]]

righe_top = "".join(
    f"<tr><td style='padding:4px 16px'><b>{r['track_name']}</b></td>"
    f"<td style='padding:4px 16px'>{r['artist_name']}</td>"
    f"<td style='padding:4px 16px; text-align:center'>{r['track_pop']}</td></tr>"
    for _, r in top10.iterrows()
)

display(HTML(f"""
<h3>🏆 Top 10 brani più popolari</h3>
<table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px'>
  <thead style='background:#ff0050; color:white'>
    <tr>
      <th style='padding:7px 16px'>Titolo</th>
      <th style='padding:7px 16px'>Artista</th>
      <th style='padding:7px 16px'>Popolarità</th>
    </tr>
  </thead>
  <tbody>{righe_top}</tbody>
</table>
"""))

# ── TOP 10 ARTISTI PIÙ PRESENTI ───────────────────────────────────────
top_artisti = df["artist_name"].value_counts().head(10)

righe_artisti = "".join(
    f"<tr><td style='padding:4px 16px'><b>{artista}</b></td>"
    f"<td style='padding:4px 16px; text-align:center'>{'🎵' * min(n, 10)} {n}</td></tr>"
    for artista, n in top_artisti.items()
)

display(HTML(f"""
<h3>📊 Artisti più presenti</h3>
<table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px'>
  <thead style='background:#ff0050; color:white'>
    <tr>
      <th style='padding:7px 16px'>Artista</th>
      <th style='padding:7px 16px'>N° brani</th>
    </tr>
  </thead>
  <tbody>{righe_artisti}</tbody>
</table>
"""))

# ── MEDIE DELLE CARATTERISTICHE AUDIO ────────────────────────────────
medie = df[["danceability", "energy", "valence", "acousticness"]].mean().round(3)

righe_medie = "".join(
    f"<tr><td style='padding:4px 16px'><b>{col}</b></td>"
    f"<td style='padding:4px 16px; text-align:center'>{val}</td></tr>"
    for col, val in medie.items()
)

display(HTML(f"""
<h3>📊 Caratteristiche audio medie dei brani virali</h3>
<table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px'>
  <thead style='background:#ff0050; color:white'>
    <tr>
      <th style='padding:7px 16px'>Caratteristica</th>
      <th style='padding:7px 16px'>Valore medio</th>
    </tr>
  </thead>
  <tbody>{righe_medie}</tbody>
</table>
"""))

# ── RICERCA INTERATTIVA ───────────────────────────────────────────────
testo_input = widgets.Text(
    value="Harry Styles",
    description="🔎 Cerca:",
    placeholder="Artista, titolo o album",
    layout=widgets.Layout(width="420px")
)
btn = widgets.Button(description="Cerca", button_style="primary", icon="search")
output = widgets.Output()

def cerca(b):
    output.clear_output()
    query = testo_input.value.strip()
    with output:
        if not query:
            print("⚠️ Inserisci un artista, titolo o album.")
            return

        q = query.lower()
        risultati = df[
            df["track_name"].str.lower().str.contains(q, na=False) |
            df["artist_name"].str.lower().str.contains(q, na=False) |
            df["album"].str.lower().str.contains(q, na=False)
        ]

        if risultati.empty:
            display(HTML(f"<span style='color:red'>❌ Nessun risultato per '<b>{query}</b>'.</span>"))
            return

        display(genera_tabella(risultati, f"Risultati per \"{query}\""))

btn.on_click(cerca)
display(HTML("<h3>🔎 Filtra per artista, titolo o album</h3>"))
display(widgets.HBox([testo_input, btn]), output)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML
from collections import defaultdict

df = pd.read_csv("discogs_2022.csv")

# ── TABELLA COMPLETA ──────────────────────────────────────────────────
def genera_tabella(dataframe, titolo=""):
    righe_html = ""
    colore_alt = False
    for _, r in dataframe.iterrows():
        bg = "#f9f9f9" if colore_alt else "#ffffff"
        colore_alt = not colore_alt
        righe_html += (
            f"<tr style='background:{bg}'>"
            f"<td style='padding:5px 12px'>{r['artista']}</td>"
            f"<td style='padding:5px 12px'>{r['album']}</td>"
            f"<td style='padding:5px 12px; text-align:center'>{r['anno']}</td>"
            f"<td style='padding:5px 12px'>{r['genere']}</td>"
            f"<td style='padding:5px 12px'>{r['stile']}</td>"
            f"</tr>"
        )
    return HTML(f"""
    <h3>🎵 {titolo}</h3>
    <p style='color:gray'>Totale: <b>{len(dataframe)}</b> vinili</p>
    <div style='max-height:500px; overflow-y:auto'>
    <table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px; width:100%'>
      <thead style='background:#1a1a2e; color:white; position:sticky; top:0'>
        <tr>
          <th style='padding:7px 12px'>Artista</th>
          <th style='padding:7px 12px'>Album</th>
          <th style='padding:7px 12px'>Anno</th>
          <th style='padding:7px 12px'>Genere</th>
          <th style='padding:7px 12px'>Stile</th>
        </tr>
      </thead>
      <tbody>{righe_html}</tbody>
    </table>
    </div>
    """)

display(genera_tabella(df, "Vinili più popolari su Discogs – USA 2022"))

# ── CONTEGGIO PER GENERE ──────────────────────────────────────────────
conteggio_genere = defaultdict(int)
for genere in df["genere"].dropna():
    for g in genere.split(", "):
        if g and g != "N/A":
            conteggio_genere[g] += 1

righe_genere = "".join(
    f"<tr><td style='padding:4px 16px'><b>{g}</b></td>"
    f"<td style='padding:4px 16px; text-align:center'>{'🎵' * min(n, 10)} {n}</td></tr>"
    for g, n in sorted(conteggio_genere.items(), key=lambda x: -x[1])
)

display(HTML(f"""
<h3>📊 Distribuzione per genere</h3>
<table border='1' cellspacing='0' style='border-collapse:collapse; font-size:13px'>
  <thead style='background:#1a1a2e; color:white'>
    <tr>
      <th style='padding:7px 16px'>Genere</th>
      <th style='padding:7px 16px'>N° vinili</th>
    </tr>
  </thead>
  <tbody>{righe_genere}</tbody>
</table>
"""))

# ── RICERCA INTERATTIVA ───────────────────────────────────────────────
testo_input = widgets.Text(
    value="Rock",
    description="🔎 Cerca:",
    placeholder="Artista, album o genere",
    layout=widgets.Layout(width="420px")
)
btn = widgets.Button(description="Cerca", button_style="primary", icon="search")
output = widgets.Output()

def cerca(b):
    output.clear_output()
    query = testo_input.value.strip()
    with output:
        if not query:
            print("⚠️ Inserisci un artista, album o genere.")
            return

        q = query.lower()
        risultati = df[
            df["artista"].str.lower().str.contains(q, na=False) |
            df["album"].str.lower().str.contains(q, na=False) |
            df["genere"].str.lower().str.contains(q, na=False) |
            df["stile"].str.lower().str.contains(q, na=False)
        ]

        if risultati.empty:
            display(HTML(f"<span style='color:red'>❌ Nessun risultato per '<b>{query}</b>'.</span>"))
            return

        display(genera_tabella(risultati, f"Risultati per \"{query}\""))

btn.on_click(cerca)
display(HTML("<h3>🔎 Filtra per artista, album o genere</h3>"))
display(widgets.HBox([testo_input, btn]), output)
```
